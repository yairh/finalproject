```python
PATH = r'D:\Documents\ITC\Projects\Data Science\data'
file = 'Data_Entry_2017.csv'

infile = os.path.join(PATH, file)
data = pd.read_csv(infile)
name = list(data.columns)
name[1] = 'Finding Label'
data.columns = name

pattern = r"|"
filter_mulitple = data['Finding Label'].str.contains(pattern)
data = data[filter_mulitple]

ind_to_keep = list()

for ind,val in enumerate(data['Finding Label']):
    m = re.findall(r'[|]',val)
    if len(m) == 0:
        ind_to_keep.append(ind)
        
data = data.loc[ind_to_keep]
```

# Creating the train and test

```python
directory = r'C:\Users\lastm\Documents\ITC\Projects\Data Science\images\images11'

df = pd.DataFrame(columns=data.columns)

for filename in os.listdir(directory):
#     print(data[data['Image Index'] == filename]['Finding Labels'])
    df = df.append(data[data['Image Index'] == filename], ignore_index=True)
    
X = df
y = df['Finding Label']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, stratify = y)

```

```python
PATH = r'C:\Users\lastm\Documents\ITC\Projects\Data Science\images\images11'

source_train = r'C:\Users\lastm\Documents\ITC\Projects\Data Science\images\train'
source_test = r'C:\Users\lastm\Documents\ITC\Projects\Data Science\images\test'
# train separation

for filename in os.listdir(PATH):
    if filename in X_train['Image Index'].values:
        dest = os.path.join(source_train, df[df['Image Index'] == filename]['Finding Label'].values[0])
        if not os.path.exists(dest):
            os.mkdir(dest)
            os.chmod(dest,0o777)
            shutil.move(os.path.join(PATH,filename), dest)
        else:
            shutil.move(os.path.join(PATH,filename), dest)
            
    if filename in X_test['Image Index'].values:
        dest = os.path.join(source_test, df[df['Image Index'] == filename]['Finding Label'].values[0])
        if not os.path.exists(dest):
            os.mkdir(dest)
            os.chmod(dest,0o777)
            shutil.move(os.path.join(PATH,filename), dest)
        else:
            shutil.move(os.path.join(PATH,filename), dest)
    else:
        continue
        
        
```
