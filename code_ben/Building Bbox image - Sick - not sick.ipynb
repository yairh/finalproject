# Cropping the images

```python
PATH = r'D:\Documents\ITC\Projects\Data Science\images\disease_and_no_finding'
```

```python
for filename in os.listdir(PATH):
    img = cv2.imread(os.path.join(PATH, filename))
    
    df_temp = df[df['Image Index'] == filename]
    x1 = math.floor(df_temp.x)
    y1 = math.floor(df_temp.y)
    x2 = math.floor(x1+ df_temp.w)
    y2 = math.floor(y1 + df_temp.h)
    
    img_crop = img[y1:y2 , x1:x2]
    
    cv2.imwrite(os.path.join(PATH, filename[:12]+'_crop.png'), img_crop)
```
