```python
df_data_entry = pd.read_csv('../../data/Data_Entry_2017.csv')
df = df_data_entry.iloc[:,:11]
df.head()
```

```python
# rows having no. of disease
df['labels_count'] = df['Finding Labels'].apply(lambda text: len(text.split('|')) 
                                                if(text == 'No Finding' or text == 'Infiltration')else 0)
df.head()
```

```python
#Open train/test lists
with open('../../data/train_val_list.txt', 'r') as train_items:
    train_list = train_items.readlines()
train_list = [item.strip() for item in train_list]

with open('../../data/test_list.txt', 'r') as test_items:
    test_list = test_items.readlines()
test_list = [item.strip() for item in test_list]
```

# Binary split: Infiltration vs No Finding

```python
# Train list
train_df = df.loc[df['Image Index'].isin(train_list) & df['labels_count'] == 1, ['Image Index','Finding Labels']]
# train_df = train_df.apply(np.random.permutation, axis=0)
train_infiltration = train_df.loc[train_df['Finding Labels'] == 'Infiltration', 'Image Index'].values
train_nf = train_df.loc[train_df['Finding Labels'] == 'No Finding', 'Image Index'].values[:len(train_infiltration)]

train_list_b = np.concatenate((train_infiltration, train_nf)) # Balanced train list
```

```python
# Test/ Val Lists
total_test_df = df.loc[df['Image Index'].isin(test_list) & df['labels_count'] == 1, ['Image Index','Finding Labels']]
# test_df = test_df.apply(np.random.permutation, axis=0)
total_infiltration = test_df.loc[test_df['Finding Labels'] == 'Infiltration', 'Image Index'].values
total_nf = test_df.loc[test_df['Finding Labels'] == 'No Finding', 'Image Index'].values[:len(total_infiltration)]

split_val_test = int(0.5 * len(total_infiltration))

val_infiltration = total_infiltration[:split_val_test]
val_nf = total_nf[:split_val_test]
val_list_b = np.concatenate((val_infiltration, val_nf))

test_infiltration = total_infiltration[split_val_test:]
test_nf = total_nf[split_val_test:]
test_list_b = np.concatenate((test_infiltration, test_nf))
```

```python
def train_test_split(train_test):
    dest = '../../imgs/images'
    if not os.path.exists(os.path.join(dest, train_test)):
        os.mkdir(os.path.join(dest, train_test))

    else:
        pass

def split_items(train_val_test_list, category):
    try:
        dest = os.path.join('../../imgs/images', category)
        for item in train_val_test_list:
            _from = os.path.join('../../imgs/images', item)
            folder = df.loc[df['Image Index'] == item, 'Finding Labels'].values[0]  # Disease
            if not os.path.exists(os.path.join(dest, folder)):
                os.mkdir(os.path.join(dest, folder))
                shutil.move(_from, os.path.join(dest, folder))
            else:
                shutil.move(_from, os.path.join(dest, folder))
    except Exception as e:
        print(e)
```

```python
# Through function
train_test_split('train')
train_test_split('val')
train_test_split('test')
        
split_items(train_list_b, 'train')
split_items(val_list_b, 'val')
split_items(test_list_b, 'test')
```
